```python
arrxr = []
archivo = open("r2.txt", "r", encoding='utf-8')
for linea in archivo:
  arrxr.append(int(linea.strip()))
archivo.close

arrxg = []
archivo = open("r2.txt", "r", encoding='utf-8')
for linea in archivo:
  arrxg.append(int(linea.strip()))
archivo.close

arrxb = []
archivo = open("r2.txt", "r", encoding='utf-8')
for linea in archivo:
  arrxb.append(int(linea.strip()))
archivo.close

arry = []
for i in range(0,256):
  arry.append(i)

```

```python
import matplotlib.pyplot as plt
import numpy as np

inicio = -2
fin=2
ancho=0.1
div=np.linspace(inicio,fin,round(1+(fin-inicio)/ancho))
plt.figure(figsize=(20,9))
plt.subplot(1,3,1)
plt.hist(arrxr,arry,div,color='red')
plt.title("Red Channel Histogram")
plt.xlabel('Intensity')
plt.ylabel('Number of Pixels')

plt.subplot(1,3,2)
plt.hist(arrxg,arry,div,color='green')
plt.title("Green Channel Histogram")
plt.xlabel('Intensity')
plt.ylabel('Number of Pixels')

plt.subplot(1,3,3)
plt.hist(arrxb,arry,div,color='blue')
plt.title("Blue Channel Histogram")
plt.xlabel('Intensity')
plt.ylabel('Number of Pixels')
```
